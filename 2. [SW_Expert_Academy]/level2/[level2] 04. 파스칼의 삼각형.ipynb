```python
T = int(input())
for tc in range(1, T+1):

    N = int(input())
    result = [[1]]

    for i in range(1, N):
        row = [1]
        for j in range(len(result[-1])-1):
            sumV = 0
            for k in range(2):
                print('j:',j,'k:',k)
                sumV += result[-1][j+k]
                print('sumV:',sumV)
            row.append(sumV)
            print('row:',sumV)
        row.append(1)
        result.append(row)
        print('result:',result, '\n')
```
