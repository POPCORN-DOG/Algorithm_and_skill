# 2. 태극문양

```python
from PIL import Image, ImageDraw

# 1. 바탕 그리기 (밝음, 순수, 평화)
w, h = 600, 400
img = Image.new('RGB', (w, h), (241,241,241))

# 2. 태극 문양 = 음양
# 1) 바탕을 만든다.
taegeuk = Image.new('RGB', (200, 200), color=(241,241,241))

# 2) 모양과 색 입히기.
draw = ImageDraw.Draw(taegeuk)
# (x, y, w, h) (0,0)에서부터 가로, 세로축(200, 200)까지의 원
# 위 빨간색(오른쪽 중간부터 0도 시작해서 아래90 왼쪽180 위270 오른쪽360 순)
draw.pieslice((0, 0, 200, 200),
             start=180, end=360, fill=(205,49,58))
# 밑 파란색
draw.pieslice((0, 0, 200, 200),
            start=0, end=180, fill=(0, 71, 160))

# 3) 타원 그리기
# 축(0,50)에서 가로, 세로축(100, 150)까지의 원
# 즉 0 - 100 = 50 - 150으로 하면 정사각형 크기
# draw.ellipse(rect(x,y,w,h), color)
draw.ellipse([0, 50, 100, 150],
            fill=(205,49,58))

draw.ellipse((100, 50, 200, 150),
            fill=(0, 71, 160))

# 4) 태극 문양 각도 맞추기
# 태극 문양의 각도는 33.69도 즉, arctan(2/3)
taegeuk = taegeuk.rotate(-33.69, 
                         fillcolor=(241,241,241))

# 5) 기존 바탕과 태극 문양 붙여주기
img.paste(taegeuk, ( 200 , 100 ))

img.show()
```

# 3. 건 (하늘, 정의)

```python
# 1) 검정 바탕 그리기
sky = Image.new('RGB', (100, 60),
                color=(14,14,14))

draw = ImageDraw.Draw(sky)

# 2) 흰색 사각형 그리기
draw.rectangle((0, 14, 100, 22),
              fill=(241,241,241))

draw.rectangle((0, 38, 100, 46),
              fill=(241,241,241))

# 3) 각도 조정하기
sky = sky.rotate(90 - 33.69,
           expand=True,
           fillcolor=(241,241,241))

img.paste(sky, (h * 22 // 96, h * 9 // 96))
img.show()
```
